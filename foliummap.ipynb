```python
import pandas as pd

df = pd.read_excel('중랑구_map.xlsx')
```

```python
import re

df_map = df.drop(columns='Unnamed: 0')
df_map
li=[]
for i in range(len(df_map)):
    a = re.search('(\w+)',df_map['시설종류명(시설유형)'][i])
    li.append(a.group())
li
df_map['유형'] = li
df_ma=df_map[df_map['유형']=='장애인']
df_ma
```

```python
import folium as g

g.map = g.Map(location = [37.585952610344, 127.07521011613], zoom_start=16)
g.map
```

```python
df_ma.reset_index(inplace = True)
```

```python
for i in range(len(df_ma)):
    marker = g.Marker([df_ma.Latitude[i], df_ma.Longitude[i]],
                     popup = df_ma.시설명[i],
                     icon = g.Icon(color = 'blue'))
    marker.add_to(g.map)
g.map
```

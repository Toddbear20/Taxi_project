```python
import pandas as pd
df_ds = pd.read_csv('장애인구별_final.csv')
df_ds = df_ds.drop(columns='Unnamed: 0')
li = []
for i in range(len(df_ds)):
    all_f = df_ds['장애인거주시설'][i]+df_ds['장애인지역사회재활시설'][i]+df_ds['장애인직업재활시설'][i]
    li.append(all_f)
df_ds['총복지시설'] = li
df_ds

```

```python
lis = []
for i in range(len(df_ds)):
    all_d= df_ds['총복지시설'][i]/df_ds['총 발달장애인 수'][i]
    lis.append(all_d)
    
df_ds['총발달장애인대비총복지시설'] = lis
mean = df_ds['총발달장애인대비총복지시설'].describe()[1]
mean
df_ds
```

```python
df_ds['복지시설부족여부'] = df_ds['총발달장애인대비총복지시설']>mean
df_ds['복지시설부족여부'].replace({True : '충분', False : '부족'}, inplace = True)
# df_ds.drop(columns = '평균지표',inplace = True)
df= df_ds.sort_values('총발달장애인대비총복지시설')
df=df.reset_index()
df=df.drop(columns='index')
df
```

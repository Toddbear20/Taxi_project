```python
import pandas as pd

df = pd.read_excel('서울시_구별_장애인유형_계.xlsx')
df['구별 총 장애인 수'] = df.sum(axis= 1)
df.columns
```

```python
df_w = pd.DataFrame(columns=['구별','구별_지체비율', '구별_뇌병변비율','구별_시각비율','구별_청각비율','구별_언어비율','구별_지적비율','구별_자폐비율','구별_정신장애비율','구별_신장비율','구별_심장비율','구별_호흡기비율','구별_간비율','구별_안면비율','구별_장루요루비율','구별_뇌전증비율'])

df_w['구별'] = df['구별']
df_w['구별_지체비율'] = df['지체_계']/df['구별 총 장애인 수']
df_w['구별_뇌병변비율'] = df['뇌병변_계']/df['구별 총 장애인 수']
df_w['구별_시각비율'] = df['시각_계']/df['구별 총 장애인 수']
df_w['구별_청각비율'] = df['청각_계']/df['구별 총 장애인 수']
df_w['구별_언어비율'] = df['언어_계']/df['구별 총 장애인 수']
df_w['구별_지적비율'] = df['지적장애_계']/df['구별 총 장애인 수']
df_w['구별_자폐비율'] = df['자폐성_계']/df['구별 총 장애인 수']
df_w['구별_정신장애비율'] = df['정신장애_계']/df['구별 총 장애인 수']
df_w['구별_신장비율'] = df['신장장애_계']/df['구별 총 장애인 수']
df_w['구별_심장비율'] = df['심장장애_계']/df['구별 총 장애인 수']
df_w['구별_호흡기비율'] = df['호흡기_계']/df['구별 총 장애인 수']
df_w['구별_간비율'] = df['간_계']/df['구별 총 장애인 수']
df_w['구별_안면비율'] = df['안면_계']/df['구별 총 장애인 수']
df_w['구별_장루요루비율'] = df['장루요루_계']/df['구별 총 장애인 수']
df_w['구별_뇌전증비율'] = df['뇌전증_계']/df['구별 총 장애인 수']



df_w = df_w.set_index('구별')
df_w
```

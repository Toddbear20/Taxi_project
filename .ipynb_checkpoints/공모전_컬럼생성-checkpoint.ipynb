## - 장애인 복지 시설

```python
facilities = pd.read_csv('./서울시_사회복지시설_목록.csv', encoding='cp949')
facilities.head()
```

```python
facilities["시설종류명(시설유형)"].unique()
```

```python
for i in range(len(facilities)): # df 의 행 수만큼 for 문 돌린다.
    fullstring = facilities.loc[i,'시설종류상세명(시설종류)'] # news 컬럼만의 각 값을 순서대로 가져온다.
    
    if search("장애인", fullstring):
        
        #장애인거주시설
        substring1 = "거주"
        substring1_2 = "공동생활"
        
        #장애인직업재활시설
        substring2 = "근로"
        substring2_1 ="직업"
        substring2_2 ="판매"
        
    
        if search(substring1, fullstring) or search(substring1_2, fullstring):
            category =  "장애인거주시설"
        elif search(substring2, fullstring) or search(substring2_1, fullstring) or search(substring2_2, fullstring):
            category =  "장애인직업재활시설"
        else:
            category =  "장애인지역사회재활시설"

    else:
        category= None

    facilities.loc[i,'장애인_시설_분류']= category
    
facilities.head(50)
```

```python
facilities = facilities[facilities.시군구명 != "서울특별시"]
```

```python
handicapped = facilities[~facilities['장애인_시설_분류'].isnull()]
```

```python
handicapped_pivot = pd.pivot_table(handicapped, 
                 index= "시군구명" ,
                 columns= "장애인_시설_분류"  ,
                 values= "시설명"    ,
                 aggfunc= "count"    )
```

## 장애인 연금 대상자이지만 어떠한 이유로 미수급자 장애인 중 시설이 필요한 장애인의 비율 

> 대상자(성인 중증 & 낮은소득) 중 시설이 필요하지만 어떠한 이유로 장애인 연금 수급을 못 받고 있는 장애인의 수는?

> 이 컬럼을 통해 장애인 생활 시설을 필요로 하지만 어떠한 이유로 혜택을 받지 못하는 취약 계층의 장애인 수를 구별로 파악 가능  

미수급자 중 시설 필요자의 수를 정확히 알 수 없지만, 최대한 근접한 수를 계산하는 방법: 

1. 구별로 장애인 연금 수급자에서 시설 수급자 비율 계산
2. 1번에서 계산한 구별 비율의 평균 비율값을 계산 
3. 구별 미수급자 수 계산 후, 평균 비율값을 곱해 시설 수급 필요자를 파악

```python
df_1= pd.read_csv('./장애인연금_장애정도별_소득계층별_수급자수.csv', encoding='cp949')
seoul_df = df_1.loc[(df_1['통계시도명'] == "서울특별시") & (df_1['소득계층구분'] == "시설수급자")]
seoul_df.drop("통계년월",axis=1, inplace=True)

seoul_df = seoul_df.drop(120)
seoul_df = seoul_df.reset_index(drop=True)
seoul_df.drop(columns=['통계시도명','장애정도','소득계층구분'], axis=1, inplace=True)
seoul_df.rename(columns = {'통계시군구명':'구별', '수급자수':'장애인연금_시설수급자수'}, inplace = True)
seoul_df
```

```python
df_combined1 = seoul_df.merge(dff, how="inner", left_on="구별", right_on="구별")
df_combined1["시설수급자평균"] = df_combined1["장애인연금_시설수급자수"]/df_combined1["수급자수"]
fac_mean = df_combined1["시설수급자평균"].mean()
fac_mean # 장애인 연금 수급자 중 시설 이용자는 약 4.2%
```

```python
dff["연금 미수급자중 시설 필요자 수"] = round((dff["대상자수"]-dff["수급자수"])*fac_mean).astype(int)
dff
```

## 심한 장애를 가진 장애인 중 장애인 연금 대상자 비율 

```python
dff["심한장애중_장애인연금_대상자비율"] = round(dff["대상자수"]/dff["심한 장애"],2)
dff
```

## 장애인 연령별

```python
df_ages = pd.read_csv('./장애인_연령별.csv')
```

```python
df_ages = df_ages[df_ages.연령별 != "소계"]
```

```python
def age_cat(x):
    if x == "9세미만":
        x = "나이<60"
    elif x == "10~19세":
        x = "나이<60"
    elif x == "20~29세" or x == "30~39세" or x == "40~49세" or x == "50~59세":
        x = "나이<60"
    else:
        x = "나이>=60"
        
    return x
        
df_ages["연령별"] = df_ages.연령별.apply(lambda x : age_cat(x)) 
```

```python
df_ages = df_ages.replace(',','', regex=True)

df_ages["인구수"] = df_ages.인구수.apply(pd.to_numeric, errors='coerce')
```

```python
df_ages= pd.pivot_table(df_ages, 
                 index= "동별"  ,
                 columns= "연령별"  ,
                 values= "인구수"    ,
                 aggfunc= "sum"    )
```

```python
df_ages_1 = df_ages.copy()
```

```python
df_ages["합"] = df_ages["나이<60"]+df_ages["나이>=60"]
```

```python
df_ages["나이<60"] = round(df_ages["나이<60"]/df_ages["합"],2)
df_ages["나이>=60"] = round(df_ages["나이>=60"]/df_ages["합"],2)
```

```python
df_ages.drop(columns=['합'], axis=1, inplace=True)
df_ages
```
